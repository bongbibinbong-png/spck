```python
load_dotenv()

google_api_key = os.getenv("GOOGLE_API_KEY")
genai.configure(api_key=google_api_key)
```

```python
model = genai.GenerativeModel(
    "gemini-2.5-flash",
    system_instruction="""
Bạn là chatbot của Shop Bống.

Nhiệm vụ:
- Tư vấn quần áo
- Gợi ý outfit
- Hỗ trợ thanh toán

Chỉ trả lời liên quan đến shop.
"""
)
```

```python
response = model.generate_content("gợi ý outfit đi học")

if response is None:
    print("Không có phản hồi")
elif hasattr(response, "text"):
    print(response.text)
else:
    print("Sai định dạng phản hồi")
```

```python
import os

files = ["chatbot.csv", "products.csv", "outfit.csv", "payments.csv"]

for f in files:
    if not os.path.exists(f):
        print(f"Thiếu file: {f}")

# load nếu đủ file
chatbot_df = pd.read_csv("chatbot.csv", encoding="utf-8")
products_df = pd.read_csv("products.csv", encoding="utf-8")
outfit_df = pd.read_csv("outfit.csv", encoding="utf-8")
payments_df = pd.read_csv("payments.csv", encoding="utf-8")

print("Load dữ liệu xong")
```

```python
chatbot_df.head()
```

```python
def chatbot_logic(prompt):

    prompt_lower = prompt.lower()

    for _, row in chatbot_df.iterrows():
        if row["question"] in prompt_lower:
            return row["answer"]

    if "áo" in prompt_lower or "quần" in prompt_lower:
        return '\n'.join([
            f"{row['name']} - {row['price']}k - {row['color']}"
            for _, row in products_df.iterrows()
        ])

    if "mặc" in prompt_lower:
        return '\n'.join([
            f"{row['style']}: {row['suggestion']}"
            for _, row in outfit_df.iterrows()
        ])

    if "ship" in prompt_lower:
        return '\n'.join([
            f"{row['method']}: {row['detail']}"
            for _, row in payments_df.iterrows()
        ])

    return "Xin lỗi, tôi chưa hiểu câu hỏi. Bạn hãy hỏi về sản phẩm hoặc shop nhé."
```
